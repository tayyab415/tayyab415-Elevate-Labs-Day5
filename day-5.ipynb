# **Task 3** – Identify relationships and trends

```python
#survival rate by gender

sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.show()

```

Observation - Females had a higher mortality rate

```python
#survival rate by Passenger Class

sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.show()
```

Observation - Passengers in first class (1) had the highest survival rate, while third class (3) passengers had significantly lower chances of survival

```python
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack', kde=True)
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

Observations 
 - Many children survived (possibly due to “women and children first” policy).
 - Majority of deaths were among adults aged 20–40.

```python
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare Paid by Survival Status')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.show()
```

Observation - Survivors paid higher fees

# **Task 4** – Plot histograms, boxplots, scatterplots

```python
# Histogram for Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=30, kde=True)
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplot for Fare by Passenger Class
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Pclass', y='Fare')
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
# Scatterplot for Age vs Fare
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived')
plt.title('Age vs Fare (Colored by Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

# **Task 5** – Write Observations for each visual

# **Observations**
- **Histogram** (Age Distribution):
The age distribution shows that most passengers were young to middle-aged adults (20-40 years), with a smaller number of children and elderly passengers, following a roughly normal distribution with a slight right skew.

- **Boxplot** (Fare by Passenger Class):
First-class passengers paid significantly higher fares with more variation in prices, while third-class fares were consistently low; there are several outliers in all classes, particularly in first class, indicating some passengers paid exceptionally high fares.

- **Scatterplot** (Age vs Fare with Survival):
Passengers who paid higher fares (typically first-class) had a better chance of survival regardless of age, while there's a dense cluster of lower-fare passengers across all age groups with mixed survival outcomes, suggesting that ticket price (and thus passenger class) was more influential for survival than age.



# **Task 6** – Provide summary of findings

# Titanic Dataset Analysis Summary

## Dataset Overview
The Titanic dataset contains passenger information including demographics, ticket information, and survival status from the tragic 1912 voyage.

## Key Findings

### 1. Demographic Patterns
- **Age Distribution**: Most passengers were young to middle-aged adults (20-40 years)
- **Gender**: More male passengers than female passengers were aboard
- **Class Distribution**: Majority of passengers traveled in third class

### 2. Survival Analysis
#### Gender Impact
- Women had significantly higher survival rates than men
- This strongly reflects the "women and children first" maritime protocol
- Female survival rate was approximately 75% while male survival rate was below 20%

#### Passenger Class Influence
- First-class passengers had the highest survival rate
- Third-class passengers had the lowest survival rate
- Clear correlation between ticket price (class) and survival chances

### 3. Economic Factors
- **Fare Distribution**:
  - First-class fares showed the highest variation and several outliers
  - Third-class fares were consistently low
  - Clear segregation of fare prices between classes

### 4. Correlations and Relationships
- Strong correlation between passenger class and survival
- Ticket fare showed positive correlation with survival chances
- Age alone wasn't a strong determinant of survival
- Higher fares (first class) generally indicated better survival chances regardless of age

## Statistical Significance
- The gender disparity in survival rates is statistically significant
- Class-based survival differences show clear patterns of social hierarchy influencing survival chances

## Conclusions
1. Being female was the strongest predictor of survival
2. Passenger class (and by extension, socioeconomic status) significantly influenced survival chances
3. The combination of being female and traveling in first class provided the highest probability of survival
4. Third-class passengers faced significant disadvantages in survival opportunities

*This analysis reveals both the social dynamics and tragic inequalities that played out during the Titanic disaster, providing insights into how factors like gender, class, and economic status influenced survival chances during this historical event.*
